```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
import statsmodels.api as sm
import seaborn as sns
import sklearn as sk
sns.set()
```

```python
data= pd.read_csv('Simple_Linear_Regression.csv')
data.describe()
```

```python
data
```

```python
#DEFINE DEPENDENT AND INDEPENDENT VARIABLE
```

```python
y=data['GPA']
x1=data['SAT']
```

```python
#EXPLORE DATA
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT', fontsize=10)
plt.ylabel('GPA', fontsize=10)
plt.show()
```

```python
#REGRESSION ITSELF
```

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
#plotting regression line on same scatter plot
```

```python
plt.scatter(x1,y)
yhat=0.0017*x1 + 0.275
fig=plt.plot(x1,yhat, lw=2, c='green', label='Regression line')
plt.xlabel('SAT', fontsize=10)
plt.ylabel('GPA', fontsize=10)
plt.show()
```

```python
#THE Ho OF THIS TEST is BETA =0
```

```python
plt.scatter(x1,y)
yhat=0.0017*x1 + 0
fig=plt.plot(x1,yhat, lw=2, c='green', label='Regression line')
plt.xlabel('SAT', fontsize=10)
plt.ylabel('GPA', fontsize=10)
plt.show()
```
